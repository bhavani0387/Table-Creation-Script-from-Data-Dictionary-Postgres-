```python
# Import required files
import pandas as pd
import os
```

```python
#Read the contents of the excel file
data = pd.read_excel('creation_script.xlsx')
data.head()
```

```python
#Drop if any duplicate records
data = data.drop_duplicates()
```

```python
data.head(100)
```

```python
data['schema'][0] == 'emp'
```

```python
tables = {}

for i in range(data.shape[0]):
    temp = data['schema'][i] + '.' + data['table'][i]
#     print(tables)
    if  temp not in tables.keys()  :
        tables[temp] = 'CREATE TABLE IF NOT EXISTS ' + temp + '( ' + os.linesep
    tables[temp] += data['column'][i] + ' ' + data['d_type'][i] + ' '
    
    if not data['length'].isna()[i] :
        tables[temp] += ' (' + str(data['length'][i]) + ') '
    
    if not data['default'].isna()[i] :
        tables[temp] += ' DEFAULT ' + str(data['default'][i]) + ' ' 
    
    if data['nullable'][i] == 'N':
        tables[temp] += ' NOT NULL ' 
    if data['cons_type'][i] == 'PK':
        tables[temp] += ' PRIMARY KEY '
    elif data['cons_type'][i] == 'U':
        tables[temp] += ' UNIQUE ' 
    elif data['cons_type'][i] == 'CHECK':
        tables[temp] += ' CHECK (' + data['condition'][i] + ') '
    elif data['cons_type'][i] == 'FK':
        tables[temp] += ' REFERENCES ' + data['condition'][i] + ' '
    tables[temp] += ', ' + os.linesep
```
